```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Spotify API istemci kimliği ve gizli anahtarını girin
client_id = 'your client id'
client_secret = 'your client secret'

# Çalma listesi URI'sini girin
playlist_uri = 'your playlist uri'
```

```python
# Spotify API'ye yetkilendirme sağlayın
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# Çalma listesindeki tüm şarkıları almak için bir döngü kullanın
playlist_features = []
offset = 0
limit = 100

while True:
    # Çalma listesindeki şarkıları alın
    results = sp.playlist_tracks(playlist_uri, offset=offset, limit=limit)
    tracks = results['items']

    if not tracks:
        break

    # Her şarkı için özellikleri alın
    for track in tracks:
        track_info = {}
        # Şarkının ID'sini alın
        track_info['ID'] = track['track']['id']
        # Şarkının adı ve sanatçılarını alın
        track_info['Name'] = track['track']['name']
        track_info['Artists'] = ', '.join([artist['name'] for artist in track['track']['artists']])

        # Şarkının özelliklerini alın
        audio_features = sp.audio_features(track['track']['uri'])
        if audio_features:
            audio_features = audio_features[0]
            for feature in ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                            'instrumentalness', 'liveness', 'valence', 'tempo']:
                track_info[feature] = audio_features.get(feature)
        playlist_features.append(track_info)

    offset += limit
```

```python
# Özellikleri içeren bir DataFrame oluşturun
df = pd.DataFrame(playlist_features)
df
```

```python
#id'leri ayırın ve gereksiz sütunları kaldırın
df_id=df[['ID']]
veri=df.drop(['ID','Name','Artists'],axis=1)
```

```python
#Standart Scaler ile veriyi scalerize edin
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
veri=sc.fit_transform(veri)
```

```python
#Dataframe oluşturun
veri=pd.DataFrame(veri)
veri.columns = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
#uzaklik isimli istenilen şarkının verilerinin tüm şarkıların verilere uzaklığını tutan bir dataframe oluşturun
uzaklik = {'danceability': [], 'energy': [], 'loudness': [], 'speechiness': [],
        'acousticness': [], 'instrumentalness': [], 'liveness': [], 'valence': [], 'tempo': []}

uzaklik = pd.DataFrame(uzaklik)
uzaklik.columns = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
#Aradığımız şarkının id'sini girin ve index değerini kaydedin
sarki_id='6E9Y4e7yjRhksdZMOU8H4D'
indeks=df.loc[df['ID']==sarki_id]
indeks=indeks.index.item()
```

```python
#Elinizde bulunan tüm şarkı verilerinin ilgili sütunları ile istenen şarkının verilerini kıyaslayın
#Uzaklık dataframe'inin ilgili satır ve sütununa atın
for j in veri:
    for i in range(len(veri)):
        uzaklik.loc[i,j] = (veri.loc[indeks,j] - veri.loc[i,j])**2
```

```python
#Toplam adında bir Dataframe oluşturun
toplam = {'id':[], 'toplam': []}
toplam = pd.DataFrame(toplam)
toplam.columns=['id','toplam']
toplam = pd.DataFrame({'id':[np.nan] * len(veri),'toplam': [0] * len(veri)})
toplam
```

```python
#uzaklık dataframeindeki satırları toplayıp toplam dataframeine atın
for i in range(len(veri)):
    a = 0
    for j in uzaklik:
        a += uzaklik.loc[i, j]
    
    toplam.loc[i, 'toplam'] += math.sqrt(a)
    toplam.loc[i,'id'] = df_id.loc[i,'ID']
```

```python
#Verileri sıralayın
toplam=toplam.sort_values('toplam')
toplam
```

```python
#karar vermek için bir grafik çizin
x=toplam['toplam']
y=range(len(x))
plt.plot(y,x)
plt.show()
```

```python
#boş bir liste oluşturup içine id'leri ekleyin
#buradaki n değerini değiştirerek şarkı sayısı ve benzerliği ile oynayabilirsiniz
n=3
liste=[]
for i in range(len(toplam)):
    if toplam.loc[i,'toplam']<n:
        liste.append(toplam.loc[i,'id'])
```
